## Distance covered prior to sub

```python
import numpy as np

injury_event = sb.events(match_id=3895302)

def distance_covered(injury_event):
    # Collect the x,y for the player at each recorded event prior to the injury
    player_movement_prior = injury_event[injury_event['player'].isin(injured_players['player']) & 
                                     injury_event['player'].map(injured_players.set_index('player')['minute']).ge(injury_event['minute'])].location

    df = pd.DataFrame(player_movement_prior.dropna().tolist(), columns=['x', 'y'])

    # Get the euclidean distance for the change in each event, where the sum is the total distance covered
    df['x_diff'] = df['x'].diff()
    df['y_diff'] = df['y'].diff()

    df['euclidean_distance'] = np.sqrt(df['x_diff']**2 + df['y_diff']**2)
    
    return df['euclidean_distance'].sum()

injury_event['distance_covered'] = distance_covered(injury_event)
final_event_data = injury_event[['match_id', 'player_id', 'player', 'substitution_outcome', 'minute', 'distance_covered']].dropna()
final_event_data = final_event_data[final_event_data['substitution_outcome'] == 'Injury']
```

```python
etl_data = final_event_data.merge(matches[['match_id', 'match_date', 'kick_off', 'home_team', 'stadium']], on='match_id')
etl_data
```
